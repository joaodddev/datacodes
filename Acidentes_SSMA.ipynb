Análise de Tendências e Insights

```python
# Criando uma função para análise de tendência
def analisar_tendencia(serie):
    x = np.arange(len(serie))
    coef = np.polyfit(x, serie, 1)
    tendencia = coef[0] * len(serie)  # Mudança total no período
    return tendencia
```

```python
# Análise de tendência mensal
tendencia_mensal = analisar_tendencia(acidentes_mes['Quantidade'])
print(f"Tendência mensal: {tendencia_mensal:.2f} (valores positivos indicam aumento)")
```

```python
# Gráfico de radar para tipos de acidente
def create_radar_chart(categories, values, title):
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    values += values[:1]
    angles += angles[:1]
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.fill(angles, values, alpha=0.25)
    ax.set_thetagrids([a * 180/np.pi for a in angles[:-1]], categories)
    plt.title(title, size=14, y=1.1)
    plt.tight_layout()
    plt.show()
```

```python
# Preparando dados para o gráfico de radar
categorias = acidentes_tipo['Tipo'].tolist()
valores = acidentes_tipo['Quantidade'].tolist()
create_radar_chart(categorias, valores, 'Distribuição de Tipos de Acidente - Safra 25/26')
```
